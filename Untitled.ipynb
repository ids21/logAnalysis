```python
import dateutil.parser
import dateutil.tz
import re
UTC_ZONE = dateutil.tz.gettz('UTC')
log_timestamp =re.compile(
    '([0-9]+)'  # day
    '[/]'
    '([A-Za-z]+)'  # month
    '[/]'
    '([0-9]+)'  # year
    '[:]'
    '([0-9]+)'  # hour
    '[:]'
    '([0-9]+)'  # minute
    '[:]'
    '([0-9]+)'  # second
    '\s+'
    '([^\s]+)'  # timezone
)
def get_datetime(timestamp_wtz):
    day, month, year, hour, minute, second, tz =\
            log_timestamp.findall(timestamp_wtz)[0]
    fmt = '%s %s %s %s:%s:%s %s' % (month, day, year, hour, minute, second, tz)
    dt = dateutil.parser.parse(fmt)
    dt = dt.astimezone(UTC_ZONE)
    return dt

data[4] = data[4].apply(get_datetime)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

groupProductsByCountAdd = basket.groupby('products',as_index=False).agg({'UserID':'count'}).rename(columns={'UserID':'Count'})


fig, ax = plt.subplots(figsize=(19, 10), subplot_kw=dict(aspect="equal"))

recipe = groupProductsByCountAdd.sort_values('Count',ascending=False)[:10]['products'].to_numpy()

top10data = groupProductsByCountAdd.sort_values('Count',ascending=False)[:10].Count;


def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d}".format(absolute)

wedges, texts,autotexts = ax.pie(top10data, wedgeprops=dict(width=0.5), startangle=-40,autopct=lambda pct: func(pct, top10data),
                                  textprops=dict(color="w"))

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/3. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

plt.setp(autotexts, size=8, weight="bold")
ax.set_title("Топ 10 товаров")

plt.show()
```

```python
apr = basket.pivot_table(index= 'UserID',columns = 'products',values='status').fillna(0)
def encode_units(x):
    if x <= 0:
        return False
    if x >= 1:
        return True
basket_sets = apr.applymap(encode_units)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,min_threshold=0.1)
rules[rules['confidence'] >= 0.6]
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fp = fpgrowth(basket_sets, min_support=0.01, use_colnames=True)
fp.sort_values(by='support',ascending=False)
fp
```

```python
rules1 = association_rules(fp,min_threshold=0.1)
rules1[rules1['confidence'] >= 0.1]
```
